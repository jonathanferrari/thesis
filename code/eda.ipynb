```python
import data
import cv2
import numpy as np
```

```python
def vid_load(file):
    cap = cv2.VideoCapture(file)
    frames = []
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frames.append(frame)
    cap.release()
    return np.array(frames)
vid_lens = []
n = len(data.files)
counter = 0
for file in data.files:
    vid = vid_load(file)
    vid_lens.append(len(vid))
    counter += 1
    print(f"{counter}/{n}", end = "\r")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = "darkgrid", palette = "tab10")
sns.histplot(vid_lens, bins = 50)
plt.title("Histogram of video lengths")
```
